```python
import pandas as pd

url = "https://raw.githubusercontent.com/jpatokal/openflights/master/data/airlines.dat"

columns = [
    "airline_id",
    "name",
    "alias",
    "iata",
    "icao",
    "callsign",
    "country",
    "active"
]

df = pd.read_csv(url, names=columns)


df_clean = df[["airline_id", "name", "iata", "icao", "callsign", "country", "active"]]

df_clean.to_csv("airlines_metadata.csv", index=False)
```
