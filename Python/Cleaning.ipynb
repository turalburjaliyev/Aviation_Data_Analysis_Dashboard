```python
# 1. SÜTUN ADLARININ STANDARTLAŞDIRILMASI
# Hamısını kiçik hərf edirik və boşluqları alt xətt (_) ilə əvəz edirik
df1.columns = df1.columns.str.lower().str.replace(' ', '_').str.replace('/', '_')
```

```python
# 2. BOŞ DƏYƏRLƏRİN TƏMİZLƏNMƏSİ
# "—" işarəsini boşluqla əvəz edirik
df1.replace(['—', ''], np.nan, inplace=True)
```

```python
def simple_extract(text, unit):
    if pd.isna(text):
        return None
    
    text_str = str(text)

    if unit == 'SUM':
        numbers = re.findall(r'(\d+)', text_str)
        if numbers:
            return float(sum(map(int, numbers)))
        return None

    parts = text_str.split('·')
    
    for part in parts:
        part = part.strip()
        
        if unit.lower() in part.lower():
            try:
                # A. Vahidin başladığı yeri tapırıq
                index = part.lower().find(unit.lower())
                
                # B. Mətni həmin yerə qədər kəsirik
                before_unit = part[:index]
                
                # C. Boşluqlara bölüb sonuncu rəqəmi götürürük
                words = before_unit.split()
                
                if words:
                    last_word = words[-1]
                    
                    # D. Təmizlik (Vergül, x, × silirik)
                    clean = last_word.replace(',', '').replace('×', '').replace('x', '')
                    
                    return float(clean)
                    
            except (ValueError, IndexError):
                continue 
                
    return None
```

```python
# Çəki (Weight)
weight_cols = ['weight_gross_weight', 'weight_max_takeoff_weight', 'weight_max_landing_weight', 'weight_max_payload']
for col in weight_cols:
    df1[f'{col}_kg'] = df1[col].apply(lambda x: simple_extract(x, 'kg'))
    df1[f'{col}_lb'] = df1[col].apply(lambda x: simple_extract(x, 'lb'))
```

```python
# Məsafə (Range)
dist_cols = ['performance_range', 'performance_landing_distance', 'performance_takeoff_distance', 'performance_ceiling']
for col in dist_cols:
    if col in df1.columns:
        df1[f'{col}_m'] = df1[col].apply(lambda x: simple_extract(x, 'm'))
        df1[f'{col}_mi'] = df1[col].apply(lambda x: simple_extract(x, 'mi'))
        df1[f'{col}_km'] = df1[col].apply(lambda x: simple_extract(x, 'km'))
        df1[f'{col}_nm'] = df1[col].apply(lambda x: simple_extract(x, 'nm'))
        df1[f'{col}_ft'] = df1[col].apply(lambda x: simple_extract(x, 'ft'))
```

```python
# Sürət (Speed)
speed_cols = ['performance_approach_speed', 'performance_cruise_speed', 'performance_max_cruise_speed', 'performance_maximum_speed']
for col in speed_cols:
    df1[f'{col}_kmh'] = df1[col].apply(lambda x: simple_extract(x, 'km/h'))
    df1[f'{col}_kt'] = df1[col].apply(lambda x: simple_extract(x, 'kt'))
```

```python
if 'info_thrust' in df1.columns:
    df1['info_thrust_lbf'] = df1['info_thrust'].apply(lambda x: simple_extract(x, 'lbf'))
    df1['info_thrust_kn'] = df1['info_thrust'].apply(lambda x: simple_extract(x, 'kN'))

if 'info_power' in df1.columns:
    df1['info_power_shp'] = df1['info_power'].apply(lambda x: simple_extract(x, 'shp'))
    df1['info_power_kw'] = df1['info_power'].apply(lambda x: simple_extract(x, 'kW'))

if 'performance_fuel_burn' in df1.columns:
    df1['performance_fuel_burn_nm_gal'] = df1['performance_fuel_burn'].apply(lambda x: simple_extract(x, 'nm/gal'))
    df1['performance_fuel_burn_kml'] = df1['performance_fuel_burn'].apply(lambda x: simple_extract(x, 'km/L'))

if 'performance_rate_of_climb' in df1.columns:
    df1['performance_rate_of_climb_ft_min'] = df1['performance_rate_of_climb'].apply(lambda x: simple_extract(x, 'ft/min'))
    df1['performance_rate_of_climb_ms'] = df1['performance_rate_of_climb'].apply(lambda x: simple_extract(x, 'm/s'))

if 'size_baggage_volume' in df1.columns:
    df1['size_baggage_volume_ft3'] = df1['size_baggage_volume'].apply(lambda x: simple_extract(x, 'ft³'))
    df1['size_baggage_volume_m3'] = df1['size_baggage_volume'].apply(lambda x: simple_extract(x, 'm³'))

if 'weight_fuel_capacity' in df1.columns:
    df1['weight_fuel_capacity_gal'] = df1['weight_fuel_capacity'].apply(lambda x: simple_extract(x, 'gal'))
    df1['weight_fuel_capacity_l'] = df1['weight_fuel_capacity'].apply(lambda x: simple_extract(x, 'L'))
    df1['weight_fuel_capacity_kg'] = df1['weight_fuel_capacity'].apply(lambda x: simple_extract(x, 'kg'))
    
```

```python
size_columns = [
    'size_cabin_height', 'size_cabin_length', 'size_cabin_width',
    'size_exterior_length', 'size_fuselage_diameter', 
    'size_tail_height', 'size_wing_span'
]
for col in size_columns:
    if col in df1.columns:
        df1[f'{col}_m'] = df1[col].apply(lambda x: simple_extract(x, 'm'))
        df1[f'{col}_ft'] = df1[col].apply(lambda x: simple_extract(x, 'ft'))
```

```python
# Ümumi sərnişin tutumu (xanadakı ən böyük rəqəm)
df1['info_seats_total'] = df1['info_seats'].apply(lambda x: simple_extract(x, 'SUM'))

# Kateqoriyalara görə ayırma
df1['info_seats_economy'] = df1['info_seats'].apply(lambda x: simple_extract(x, 'Economy'))
df1['info_seats_business'] = df1['info_seats'].apply(lambda x: simple_extract(x, 'Business'))
df1['info_seats_first'] = df1['info_seats'].apply(lambda x: simple_extract(x, 'First'))
```

```python
# 1. df_basic: Əsas Məlumatlar və Oturacaqlar
basic_cols = [
    'name', 'url', 'info_avionics', 'info_engine', 'info_engine_type', 
    'info_wing_tips', 'info_seats_total', 'info_seats_economy', 
    'info_seats_business', 'info_seats_first'
]

# 2. df_range_altitude: Məsafə və Hündürlük
range_alt_cols = [
    'name', 'performance_range_km', 'performance_range_nm', 'performance_range_mi', 
    'performance_landing_distance_m', 'performance_landing_distance_ft', 
    'performance_takeoff_distance_m', 'performance_takeoff_distance_ft', 
    'performance_ceiling_m', 'performance_ceiling_ft'
]

# 3. df_speed: Sürətlər
speed_cols = [
    'name', 'performance_approach_speed_kmh', 'performance_approach_speed_kt', 
    'performance_max_cruise_speed_kmh', 'performance_max_cruise_speed_kt'
]

# 4. df_engine_metrics: Mühərrik Gücü və Effektivlik
engine_perf_cols = [
    'name', 'info_thrust_lbf', 'info_thrust_kn', 'info_power_shp', 'info_power_kw', 
    'performance_fuel_burn_nm_gal', 'performance_fuel_burn_kml', 
    'performance_rate_of_climb_ft_min', 'performance_rate_of_climb_ms'
]

# 5. df_weights: Çəki və Yanacaq Tutumu (Tam siyahı)
weight_cols = [
    'name', 'weight_max_takeoff_weight_kg', 'weight_max_takeoff_weight_lb', 
    'weight_max_landing_weight_kg', 'weight_max_landing_weight_lb', 'weight_max_payload_kg', 
    'weight_max_payload_lb', 'weight_fuel_capacity_gal', 'weight_fuel_capacity_l', 'weight_fuel_capacity_kg'
]

# 6. df_dimensions: Fiziki Ölçülər (Tam siyahı)
dim_cols = [
    'name', 'size_baggage_volume_ft3', 'size_baggage_volume_m3', 'size_cabin_height_m', 
    'size_cabin_height_ft', 'size_cabin_length_m', 'size_cabin_length_ft', 'size_cabin_width_m', 
    'size_cabin_width_ft', 'size_exterior_length_m', 'size_exterior_length_ft', 'size_fuselage_diameter_m', 
    'size_fuselage_diameter_ft', 'size_tail_height_m', 'size_tail_height_ft', 'size_wing_span_m', 'size_wing_span_ft'
]

df_basic_specs = df1[[c for c in basic_cols if c in df1.columns]].copy()
df_range_altitude_specs = df1[[c for c in range_alt_cols if c in df1.columns]].copy()
df_speed_specs = df1[[c for c in speed_cols if c in df1.columns]].copy()
df_engine_metrics_specs = df1[[c for c in engine_perf_cols if c in df1.columns]].copy()
df_weights_specs = df1[[c for c in weight_cols if c in df1.columns]].copy()
df_dimensions_specs = df1[[c for c in dim_cols if c in df1.columns]].copy()
```
